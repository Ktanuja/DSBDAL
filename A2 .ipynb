```python
## Fill null values by 0 or 1
df2 = df.fillna(value = 0)
df2
```

```python
df2.isnull().sum().sum()
#now there is no null values
```

```python
# filling null value by previous value

df3 = df.fillna(method = 'pad')
df3
```

```python
# filling null value by next-row value

df4 = df.fillna(method = 'bfill')
df4
```

```python
# filling null value by next-column value

df5 = df.fillna(method = 'pad', axis=1)
df5
```

```python
# filling null value by differ value

df6 = df.fillna({'Roll No':'1',
              'FE SGPA':'00.00'})
df6
```

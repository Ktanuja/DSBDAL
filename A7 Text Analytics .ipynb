```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
from nltk import sent_tokenize
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
text='Real madrid is set to win the UCL for the season . Benzema might win Balon dor . Salah might be the runner up'
```

**Tokenization**

breaking down a sequence of text into smaller units called tokens. These tokens can be individual words, characters, or subwords, depending on the context and requirements of the analysis.

Types:Word, character, subword

```python
w1 = nltk.word_tokenize(text)
print(w1)
```

```python
w2 = nltk.sent_tokenize(text)
print(w2)
```

```python
w1 = nltk.wordpunct_tokenize(text)
print(w1)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import LancasterStemmer
```

```python
stem=[]
for i in w1:
  ps = PorterStemmer()
  stem_word= ps.stem(i)
  stem.append(stem_word)
print(stem)
```

**Lemmatization**

reducing words to their base or root form, known as the lemma

process, different forms of a word are transformed into a single common form

 stemming is a simpler and faster approach that reduces words to their basic form without considering context or part of speech. Lemmatization, on the other hand, provides more accurate results by considering the context and POS of words,

```python
import nltk
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in stem])
print(lemmatized_output)
```

```python
leme=[]
for i in stem:
  lemetized_word=lemmatizer.lemmatize(i)
  leme.append(lemetized_word)
print(leme)
```

 Part of Speech Tagging

 assigns a specific grammatical category or part of speech to each word in a given sentence or text

  nouns, verbs, adjectives, adverbs, pronouns, prepositions, conjunctions, and interjections

```python
print(nltk.pos_tag(leme))
```

Stop word

do not carry significant meaning or contribute to the understanding of the content.

```python
sw_nltk = stopwords.words('english')
print(sw_nltk)
```

```python
words = [word for word in text.split() if word.lower() not in sw_nltk]
new_text = " ".join(words)
print(new_text)
```
